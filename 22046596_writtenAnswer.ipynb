# Question 3.iv. Express query in relational algebra
I am using **prescriptionID** 'PCT004' as an example. The relational alogebra for it is below:
- Step 1: Call PCT004 a table to show this Catersian product between 2 tables whose values are filtered.  $$PCT004 =\sigma_{pct.prescriptionID = 'PCT004'}(prescription) \times \sigma_{ph.prescriptionID = 'PCT004'}(prescriptionHerb)$$
  - $\sigma_{pct.prescriptionID = 'PCT004'}(prescription)$ table will only have 1 row and all columns from table **prescription** because prescriptionID here is primary key
  - $\sigma_{ph.prescriptionID = 'PCT004'}(prescriptionHerb)$ table have multiple (let say 3) rows and all columns from table **prescriptionHerb**
  - Therefore, after Catersian product, PCT004 table will have 3 rows and all columns that are combined from 2 original tables
- Step 2: Again, using Catersian product for all PCT004 table and **herb** table into **Summary** table $$ Summary = PCT004 \times herb$$
- Step 3: Using selection algebra to filter where herbID is matching in **Summary** table $$SummaryFiltered = \sigma_{PCT004.herbID = h.herbID}(Summary)$$
- Step 4: Finally, projection (choosing) only some features (columns) that we want to see $$\prod_{h.herbName, ph.quantity, h.price} (SummaryFiltered)$$

In short, alias of 3 tables: **presciption** (pct), **prescriptionHerb** (ph), and **herb** (h), we have $$\prod_{h.herbName, ph.quantity, h.price} (\sigma_{PCT004.herbID = h.herbID}(PCT004 \times herb))$$ or $$\prod_{h.herbName, ph.quantity, h.price} (PCT004\bowtie_{ph.herbID = h.herbID}(herb))$$ with $$PCT004 =\sigma_{pct.prescriptionID = 'PCT004'}(prescription) \times \sigma_{ph.prescriptionID = 'PCT004'}(prescriptionHerb)$$
